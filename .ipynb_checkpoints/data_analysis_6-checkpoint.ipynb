```python
import pandas as pd
uselog = pd.read_csv('data/use_log.csv')
print(len(uselog))
uselog.head()
```

```python
customer = pd.read_csv('data/customer_gym.csv')
print(len(customer))

#end_date の値がNaN以外の場所を表示、どのようなデータが入っているか確認　→　月末のデータが入っている
#customer.loc[~customer['end_date'].isnull()]

#end_dataがNaN　の行は is_deletedの値がすべて1になっている
```

```python
class_master = pd.read_csv('data/class_gym.csv')
print(len(class_master))
class_master.head()
```

```python
campaign_master = pd.read_csv('data/campaign_gym.csv')
print(len(campaign_master))
campaign_master.head()
```

### 会員データを整形する

```python
#customerデータフレームに　classデータフレームを結合する
customer_join = pd.merge(
    customer, class_master,
    on='class',#二つのデータフレームに共通する列名
    how='left' #customerを主軸に結合する
)

#上記で結合したデータフレームに campaignデータフレームを結合する
customer_join = pd.merge(
    customer_join, campaign_master,
    on='campaign_id',#二つのデータフレームに共通するフレーム
    how='left'       #customerを主軸に結合する
)
customer_join
```

### データ検算をおこなう

```python
#結合前の会員データ数を表示
print(len(customer))

#結合後の会員データ数を表示
print(len(customer_join))
```

```python
#等しくなったのでOK
```

```python
customer.isnull().sum()
```

```python
customer_join.isnull().sum()
```

```python
# .isnll().sum()　で結合前と結合後の欠損値を比べたが変化はないのでOK
```

### 会員データの基礎集計をおこなう

# データ検算のためのデータ数の確認など、何度も同じ処理を行う場合、『 関数 』 として処理内容をひとまとめにして、それを利用するとスマート

```python
#データフレームのデータ数を計算してくれる関数
def cal_number(df):
    num = 0
    for i in range(len(df)):
        num += df[i]
    return num
```

```python
#class_name・・・会員区分別にcustomer_idの数＝会員人数をカウントする
by_class = customer_join.groupby("class_name").count()["customer_id"]
#オールタイムの人数が最も多い。他の二つよりも2倍ほど多い
by_class
```

```python
#簡易グラフ化
by_class.plot.bar();
#日本語は表示できない　Warningが出る　もう一度実行すると消える
#きちんとグラフを作成したい場合は　Matplotlibライブラリを利用。ただし日本語設定が必要
```

```python
#データ検算
print(len(customer))

#上記会員区別人数の計算
#by_class[0] + by_class[1] + by_class[2]

#上記計算を関数cal_numberを使っておこなう
cal_number(by_class)
```

```python
#campaing_name・・・キャンペーン区分別にcustomer_idの数＝会員人数をカウントする
by_campaign = customer_join.groupby("campaign_name").count()["customer_id"]
#通常のキャンペーン区分が最も多い。他よりも4〜7倍多い。年中キャンペーンを実施しているわけではなさそう。

print(by_campaign)
by_campaign.plot.bar();
```

```python
#データ検算
print(len(customer))

#これでも計算できるが下の関数を利用した方がスマート
#by_campaign[0] + by_campaign[1] + by_campaign[2]
cal_number(by_campaign)
```

```python
#gender・・・性別ごとにcustomer_idの数＝会員人数をカウントする
by_gender = customer_join.groupby("gender").count()["customer_id"]
#男性の方がやや多い
print(by_gender)
by_gender.plot.bar();
```

```python
#データ検算
print(len(customer))

cal_number(by_gender)
```

# 3．利用履歴データを集計しよう
#### customer_join の　customer_id列を　小文字から大文字に変えないとうまく結合できない　NaNがはいってしまう

```python
customer_join['customer_id'] = customer_join['customer_id'].str.upper()
customer_use = pd.merge(
    uselog, customer_join,
    on='customer_id',
    how='left'
)
customer_use
```

```python
uselog["usedate"] = pd.to_datetime(uselog["usedate"])
uselog["年月"] = uselog["usedate"].dt.strftime("%Y/%m")
uselog
```

```python
uselog_months = uselog.groupby(["年月","customer_id"],as_index=False).count()
uselog_months.rename(columns={"log_id":"count"}, inplace=True)
del uselog_months["usedate"]
uselog_months.head()
```

```python
uselog_customer = uselog_months.groupby("customer_id").agg(["mean", "median", "max", "min" ])["count"]
uselog_customer.head()
```

```python
uselog_customer = uselog_customer.reset_index(drop=False)
uselog_customer.head()
```

### 利用履歴データから定期利用フラグを作成しよう

```python
uselog["weekday"] = uselog["usedate"].dt.weekday
uselog_weekday = uselog.groupby(
    ["customer_id","年月","weekday"], 
    as_index=False
).count()[
    ["customer_id","年月", "weekday","log_id"]
]

uselog_weekday.rename(columns={"log_id":"count"}, inplace=True)
uselog_weekday.head()
```

```python
uselog_weekday = uselog_weekday.groupby("customer_id",as_index=False).max()[["customer_id", "count"]]
uselog_weekday["routine_flg"] = 0


uselog_weekday["routine_flg"] = uselog_weekday["routine_flg"].where(uselog_weekday["count"] < 4, 1)
uselog_weekday.head()
```

### 会員データと利用履歴データを結合しよう

```python
customer_join = pd.merge(customer_join, uselog_customer, on="customer_id", how="left")
customer_join = pd.merge(customer_join, uselog_weekday[["customer_id", "routine_flg"]], on="customer_id", how="left")
customer_join.head()
```

### 会員期間を計算しよう

```python
from dateutil.relativedelta import relativedelta
customer_join["calc_date"] = customer_join["end_date"]
customer_join["calc_date"] = customer_join["calc_date"].fillna(pd.to_datetime("20190430"))
customer_join["membership_period"] = 0
for i in range(len(customer_join)):
    delta = relativedelta(customer_join["calc_date"].iloc[i], customer_join["start_date"].iloc[i])
    customer_join["membership_period"].iloc[i] = delta.years*12 + delta.months
customer_join.head()
```
